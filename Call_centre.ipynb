```python
#part 2
def p2(get_week):
    #Total Number of Calls in week
    print('Total Number of Calls in week',get_week,'=',df.week.value_counts()[get_week])
    #Calls Answered
    print('Total Number of Calls answered in week',get_week,'=',df[(df.week==get_week)&(df.Ans=='Y')]['Ans'].count())
    #Avg Speed of Answer 
    print('Average Speed of Answer in week',get_week,'=',df[(df.week==get_week)&(df.Ans=='Y')]['Speed of Answer'].mean())
    #Calls Abandoned
    print('Percentage of calls abandonded in week',get_week,'=',(df[(df.week==get_week)&(df.Ans=='N')]['Ans'].count()/df.week.value_counts()[get_week])*100)
    #Satisfaction overall 
    print('Satisfaction overall in week ',get_week,'=',df[(df.week==get_week)&(df.Resolved=='Y')]['Resolved'].count())
    #2.Satisfaction less than equal to 3
    print('Satisfaction less than 3 in week',get_week,'=',df[(df.week==get_week)&(df['Satisfaction rating']<=3)]['Satisfaction rating'].count())
```

```python
get_week=int(input())
p2(get_week)
```

```python
#part 3
def p3(get_name):
    #3.Total number of calls
    print('Total Number of Calls of',get_name,'=',df.Agent.value_counts()[get_name])
    #3.Total number of calls answered
    print('Number calls answered by',get_name,'=',df[(df.Agent==get_name)&(df.Ans=='Y')]['Ans'].count())
    #3.Avg Speed of Answer
    print('Average Speed of Answer by',get_name,'=',df[(df.Agent==get_name)&(df.Ans=='Y')]['Speed of Answer'].mean())
    #3.Call Resolution %
    print('Percentage of call Resolution by',get_name,'=',(df[(df.Agent==get_name)&(df.Resolved=='Y')]['Resolved'].count()/df.Agent.value_counts()[get_name])*100)
    #3.Call Resolved
    print('Number of calls Resolved by',get_name,'=',(df[(df.Agent==get_name)&(df.Resolved=='Y')]['Resolved'].count()))
```

```python
get_name=input()
p3(get_name)
```

```python
#part 4
def p4(get_name,get_weekday):
    #3.Total number of calls
    print('Number of calls by ',get_name ,'on',get_weekday,'th Weekday =',df[(df.Agent==get_name)&(df.weekday==get_weekday)]['Agent'].count())
    #3.Total number of calls answered
    print('Number of calls answered by ',get_name ,'on',get_weekday,'th Weekday =',df[(df.Agent==get_name)&(df.Ans=='Y')&(df.weekday==get_weekday)]['Ans'].count())
    #3.Avg Speed of Answer
    print('Average Speed of Calls answered by ',get_name ,'on',get_weekday,'th Weekday =',df[(df.Agent==get_name)&(df.Ans=='Y')&(df.weekday==get_weekday)]['Speed of Answer'].mean())
    #3.Call Resolution %
    print('Percentage of call Resolution by',get_name ,'on',get_weekday,'th Weekday =',(df[(df.Agent==get_name)&(df.Resolved=='Y')&(df.weekday==get_weekday)]['Resolved'].count()/df[(df.Agent==get_name)&(df.weekday==get_weekday)]['Agent'].count()*100))
    #3.Call Resolved
    print('Number of calls Resolved by',get_name ,'on',get_weekday,'th Weekday =',(df[(df.Agent==get_name)&(df.Resolved=='Y')&(df.weekday==get_weekday)]['Resolved'].count()))
```

```python
get_name=input()
get_weekday=int(input())
p4(get_name,get_weekday)
```

```python
#Part 5
def p5(get_dept):
    #Total number of calls
    print('Total Number of Calls of',get_dept,'Department =',df.Department.value_counts()[get_dept])
    #Total number of calls answered
    print('Number calls answered by',get_dept,'Department =',df[(df.Department==get_dept)&(df.Ans=='Y')]['Ans'].count())
    #Avg Speed of Answer
    print('Average Speed of Answer by',get_dept,'Department =',df[(df.Department==get_dept)&(df.Ans=='Y')]['Speed of Answer'].mean())
    #Call Resolution %
    print('Percentage of call Resolution by',get_dept,'Department =',(df[(df.Department==get_dept)&(df.Resolved=='Y')]['Resolved'].count()/df.Department.value_counts()[get_dept])*100)
    #Call Resolved
    print('Number of calls Resolved by',get_dept,'Department =',(df[(df.Department==get_dept)&(df.Resolved=='Y')]['Resolved'].count()))
```

```python
#Part 6
def p6(get_dept):
    #Total number of calls
    print('Total Number of Calls of',get_dept,'Department =',df.Department.value_counts()[get_dept])
    #Total number of calls answered
    print('Number calls answered by',get_dept,'Department =',df[(df.Department==get_dept)&(df.Ans=='Y')]['Ans'].count())
    # % Calls Abandoned
    print('Percentage of calls abandonded by',get_dept,'Department =',(df[(df.Department==get_dept)&(df.Ans=='N')]['Ans'].count()/df.Department.value_counts()[get_dept])*100)
    # % SLA Limit
    sla=(df[(df.Department==get_dept)&(df.Ans=='N')]['Ans'].count()/df.Department.value_counts()[get_dept])*100
    print('SLA Limit =',sla)
    if sla>=20 :
        print('SLA Breached: Yes')
    else :
        print('SLA Breached: No')
```

```python
get_dept=input()
p6(get_dept)
```
